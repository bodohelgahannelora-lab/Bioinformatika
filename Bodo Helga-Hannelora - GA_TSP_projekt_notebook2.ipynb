## Távolságfüggvények implementálása

### Subtask:
Implementáld a `distance` függvényt az euklideszi távolság kiszámítására két város között, valamint a `build_distance_matrix` függvényt az NxN távolságmátrix felépítésére.


**Reasoning**:
I need to implement the `distance` and `build_distance_matrix` functions as specified in the subtask. The `distance` function will calculate the Euclidean distance between two cities using their 'x' and 'y' coordinates. The `build_distance_matrix` function will then use this `distance` function to populate an NxN matrix with all pairwise distances between cities.



```python
import math

# TODO: távolság két város között
def distance(city_a, city_b):
    """Számold ki a (city_a, city_b) távolságát (pl. euklideszi)."""
    dx = city_a['x'] - city_b['x']
    dy = city_a['y'] - city_b['y']
    return math.sqrt(dx*dx + dy*dy)


# TODO: távolságmátrix építése
def build_distance_matrix(cities):
    """Adja vissza az NxN-es távolságmátrixot a megadott városlistára."""
    n_cities = len(cities)
    dist_matrix = [[0.0] * n_cities for _ in range(n_cities)]
    for i in range(n_cities):
        for j in range(i, n_cities):
            d = distance(cities[i], cities[j])
            dist_matrix[i][j] = d
            dist_matrix[j][i] = d  # A mátrix szimmetrikus
    return dist_matrix


dist_matrix = build_distance_matrix(cities)
len(dist_matrix), len(dist_matrix[0])  # ellenőrzés
```

## Tour hossz és fitness függvény implementálása

### Subtask:
Implementáld a `tour_length` függvényt egy tour hosszának kiszámítására a távolságmátrix alapján, valamint a `fitness` függvényt, amely a tour hosszát maximalizálandó értékre alakítja át.


**Reasoning**:
I need to implement the `tour_length` and `fitness` functions based on the instructions. The `tour_length` function will calculate the total distance of a given tour, including the return to the starting city, using the `dist_matrix`. The `fitness` function will then convert this length into a value that should be maximized, typically by returning the negative of the tour length.



```python
# TODO: tour hossza
def tour_length(tour, dist_matrix):
    """Számold ki a tour hosszát, feltételezve, hogy visszatér az első városba."""
    total_length = 0
    num_cities_in_tour = len(tour)
    for i in range(num_cities_in_tour):
        # Get current city index and next city index
        current_city_idx = tour[i]
        next_city_idx = tour[(i + 1) % num_cities_in_tour]  # % for returning to the start
        total_length += dist_matrix[current_city_idx][next_city_idx]
    return total_length


# TODO: fitness függvény
def fitness(tour, dist_matrix):
    """Alakítsd a tour hosszát olyan értékké, amit maximalizálni szeretnénk."""
    # pl. return -tour_length(tour, dist_matrix)
    return -tour_length(tour, dist_matrix)
```

## Mutációs operátorok implementálása

### Subtask:
Implementáld a `swap_mutation` és az `inversion_mutation` függvényeket a `AOuDhVK8wdq7` cellában a megadott feladatleírás szerint.


**Reasoning**:
I need to implement the `swap_mutation` and `inversion_mutation` functions in the specified code cell. These functions will generate mutated tours by swapping two positions or inverting a segment, respectively, ensuring they operate on copies of the input tour.



```python
import random

# TODO: swap mutáció
def swap_mutation(tour):
    """Válassz ki két pozíciót és cseréld meg őket egy másolatban."""
    mutated_tour = list(tour) # Működjünk a tour másolatán
    idx1, idx2 = random.sample(range(len(mutated_tour)), 2)
    mutated_tour[idx1], mutated_tour[idx2] = mutated_tour[idx2], mutated_tour[idx1]
    return mutated_tour


# TODO: inversion mutáció
def inversion_mutation(tour):
    """Válassz ki két indexet i<j, és fordítsd meg a tour[i:j+1] szakaszt."""
    mutated_tour = list(tour) # Működjünk a tour másolatán
    n_cities = len(mutated_tour)
    idx1, idx2 = sorted(random.sample(range(n_cities), 2))

    # Fordítsuk meg a szakaszt az idx1 és idx2 között
    mutated_tour[idx1 : idx2 + 1] = mutated_tour[idx1 : idx2 + 1][::-1]
    return mutated_tour
```

## Crossover operátor implementálása

### Subtask:
Implementálj egy TSP-kompatibilis crossover operátort (pl. OX vagy PMX) a `crossover` függvényben a `zelRlC4C4wdq7` cellában, biztosítva, hogy az utódok érvényes permutációk legyenek.


**Reasoning**:
I need to implement the `crossover` function using the Ordered Crossover (OX) method as specified in the subtask instructions. This involves selecting a segment from one parent, and then filling the remaining slots in the child with cities from the other parent in a cyclical order, ensuring no duplicates and preserving the relative order of the remaining cities.



```python
import random

# TODO: OX vagy PMX crossover implementációja
def crossover(parent1, parent2, method="OX"):
    """TSP kompatibilis crossover. Visszaadhat 1 vagy 2 utódot is."""
    n_cities = len(parent1)

    if method == "OX": # Ordered Crossover
        # Válassz ki két véletlen vágási pontot
        # sorted ensures start_idx < end_idx
        start_idx, end_idx = sorted(random.sample(range(n_cities), 2))

        child1 = [-1] * n_cities
        child2 = [-1] * n_cities

        # 1. Helyezd át a kijelölt szegmenst az első szülőről az első utódba
        child1[start_idx : end_idx + 1] = parent1[start_idx : end_idx + 1]
        # 2. Helyezd át a kijelölt szegmenst a második szülőről a második utódba
        child2[start_idx : end_idx + 1] = parent2[start_idx : end_idx + 1]

        # Segédfüggvény a hiányzó városok kitöltéséhez
        def fill_child(target_child, source_parent, segment_from_other_parent):
            segment_cities_set = set(segment_from_other_parent)
            # Hozd létre a source_parent-ből azokat a városokat, amelyek nem szerepelnek a segment_cities_set-ben
            # megtartva a sorrendjüket
            remaining_cities_in_order = [
                city for city in source_parent if city not in segment_cities_set
            ]

            # Töltsd ki az üres helyeket (ahol -1 van) a remaining_cities_in_order elemeivel
            # kezdve az end_idx utáni pozíciótól, ciklikusan
            fill_pos = (end_idx + 1) % n_cities
            for city in remaining_cities_in_order:
                while target_child[fill_pos] != -1:
                    fill_pos = (fill_pos + 1) % n_cities
                target_child[fill_pos] = city

        # Töltsd ki child1-et parent2 hiányzó elemeivel
        fill_child(child1, parent2, parent1[start_idx : end_idx + 1])
        # Töltsd ki child2-t parent1 hiányzó elemeivel
        fill_child(child2, parent1, parent2[start_idx : end_idx + 1])

        return child1, child2
    else:
        raise ValueError(f"Ismeretlen crossover módszer: {method}")
```

## Szelekciós operátor implementálása

### Subtask:
Implementáld a `select_parent` függvényt egy szülő kiválasztására a populációból a fitneszek alapján, például tournament szelekcióval a `EoN-G79hwdq7` cellában.


**Reasoning**:
I need to implement the `select_parent` function using the tournament selection method as specified in the instructions. This involves randomly selecting `k` individuals, comparing their fitness values, and returning the one with the highest fitness.



```python
import random

# TODO: szelekció – pl. tournament selection
def select_parent(population, fitness_values, method="tournament", k=3):
    """Válassz ki egy szülőt a population-ből a fitness_values alapján."""
    if method == "tournament":
        n_individuals = len(population)
        # Válassz ki k egyedi indexet a populációból
        tournament_indices = random.sample(range(n_individuals), k)

        # Gyűjtsd össze a kiválasztott tourokat és fitneszértékeket
        tournament_individuals = [population[i] for i in tournament_indices]
        tournament_fitnesses = [fitness_values[i] for i in tournament_indices]

        # Keresd meg a legjobb fitneszű egyedet a tournament csoportban
        # Mivel a fitness függvényt maximalizáljuk (negatív hosszat ad vissza),
        # a max() függvény adja vissza a "legjobbat" (legkisebb hosszat).
        best_fitness_in_tournament = max(tournament_fitnesses)
        best_individual_index = tournament_fitnesses.index(best_fitness_in_tournament)

        return tournament_individuals[best_individual_index]
    else:
        raise ValueError(f"Ismeretlen szelekciós módszer: {method}")
```

## GA fő ciklus implementálása

### Subtask:
Implementáld a `run_ga` függvényt, amely tartalmazza a populáció inicializálását, a fitneszek számítását, szülők kiválasztását, crossover és mutáció alkalmazását, valamint az elitizmust a `cYmXxdrDwdq8` cellában. Ez a függvény adja vissza a legjobb tour-t, a hozzá tartozó hosszt és a konvergenciagörbéket.


**Reasoning**:
I need to implement the `run_ga` function as per the instructions. This function will orchestrate the entire genetic algorithm process, including population initialization, fitness evaluation, selection, crossover, mutation, and elitism, and track the progress over generations. The implementation will follow the detailed steps provided in the instructions.



```python
import random
import numpy as np

# TODO: GA fő ciklus
def run_ga(
    cities,
    pop_size=50,
    n_generations=300,
    crossover_rate=0.9,
    mutation_rate=0.1,
    mutation_mode="mixed",  # "swap", "inversion", "mixed"
    elitism=1
):
    """Futtasd a GA-t TSP-re, és add vissza a legjobb megoldást és statisztikákat."""
    n_cities = len(cities)
    dist_matrix = build_distance_matrix(cities)

    # Inicializálás
    population = init_population(pop_size, n_cities)

    best_tour = None
    best_length = float('inf')

    # Történet rögzítése
    history = {
        "best_lengths": [],
        "avg_lengths": []
    }

    for gen in range(n_generations):
        # Fitneszek számítása
        fitness_values = [fitness(tour, dist_matrix) for tour in population]
        tour_lengths = [-f for f in fitness_values] # Valódi tour hosszak

        # Best tour frissítése
        current_best_fitness = max(fitness_values)
        current_best_tour_idx = fitness_values.index(current_best_fitness)
        current_best_tour_length = tour_lengths[current_best_tour_idx]

        if current_best_tour_length < best_length:
            best_length = current_best_tour_length
            best_tour = population[current_best_tour_idx]

        # Történet mentése
        history["best_lengths"].append(current_best_tour_length)
        history["avg_lengths"].append(np.mean(tour_lengths))

        new_population = []

        # Elitizmus
        # Rendezés fitnesz szerint csökkenő sorrendben (legjobb fitnesz előre)
        sorted_population_indices = np.argsort(fitness_values)[::-1]
        for i in range(elitism):
            new_population.append(population[sorted_population_indices[i]])

        # Új populáció generálása
        while len(new_population) < pop_size:
            parent1 = select_parent(population, fitness_values, method="tournament")
            parent2 = select_parent(population, fitness_values, method="tournament")

            child1, child2 = parent1, parent2 # Alapértelmezésben, ha nincs crossover

            if random.random() < crossover_rate:
                child1, child2 = crossover(parent1, parent2, method="OX")

            # Mutáció
            if random.random() < mutation_rate:
                if mutation_mode == "swap":
                    child1 = swap_mutation(child1)
                elif mutation_mode == "inversion":
                    child1 = inversion_mutation(child1)
                elif mutation_mode == "mixed":
                    if random.random() < 0.5:
                        child1 = swap_mutation(child1)
                    else:
                        child1 = inversion_mutation(child1)

            if random.random() < mutation_rate:
                if mutation_mode == "swap":
                    child2 = swap_mutation(child2)
                elif mutation_mode == "inversion":
                    child2 = inversion_mutation(child2)
                elif mutation_mode == "mixed":
                    if random.random() < 0.5:
                        child2 = swap_mutation(child2)
                    else:
                        child2 = inversion_mutation(child2)

            new_population.append(child1)
            if len(new_population) < pop_size: # Csak akkor adjuk hozzá a második gyermeket, ha még van hely
                new_population.append(child2)

        population = new_population

    # A generációk végén a legjobb tour hossza
    final_fitness_values = [fitness(tour, dist_matrix) for tour in population]
    final_tour_lengths = [-f for f in final_fitness_values]
    current_best_fitness = max(final_fitness_values)
    current_best_tour_idx = final_fitness_values.index(current_best_fitness)
    current_best_tour_length = final_tour_lengths[current_best_tour_idx]

    if current_best_tour_length < best_length:
        best_length = current_best_tour_length
        best_tour = population[current_best_tour_idx]

    return best_tour, best_length, history


# TODO: paraméterezett futtatás (példa)
# best_tour, best_length, history = run_ga(cities)
```

The `plot_tour` function will visualize the best tour by plotting cities as points and connecting them in the tour order, including returning to the start. The `plot_convergence` function will plot the best and average tour lengths over generations from the provided history.



```python
import matplotlib.pyplot as plt

# TODO: rajzold ki a legjobb tour-t
def plot_tour(cities, tour, title="Best tour"):
    """Ábrázolja a legjobb tour-t a síkon."""
    plt.figure(figsize=(10, 8))
    # Városok kirajzolása
    x_coords = [cities[i]['x'] for i in tour]
    y_coords = [cities[i]['y'] for i in tour]

    plt.scatter(x_coords, y_coords, color='red', zorder=5) # Városok pontokként

    # Élek összekötése
    for i in range(len(tour)):
        start_city_idx = tour[i]
        end_city_idx = tour[(i + 1) % len(tour)] # Vissza az elsőhöz

        start_x = cities[start_city_idx]['x']
        start_y = cities[start_city_idx]['y']
        end_x = cities[end_city_idx]['x']
        end_y = cities[end_city_idx]['y']

        plt.plot([start_x, end_x], [start_y, end_y], color='blue', linestyle='-', linewidth=1)

    # Városnevek hozzáadása (opcionális)
    for i, city_data in enumerate(cities):
        plt.annotate(city_data['name'], (city_data['x'], city_data['y']), textcoords="offset points", xytext=(5,5), ha='center')

    plt.title(title)
    plt.xlabel("X koordináta")
    plt.ylabel("Y koordináta")
    plt.grid(True)
    plt.show()


# TODO: konvergenciagörbe rajzolása
def plot_convergence(history):
    """Rajzolja ki a konvergenciagörbét."""
    plt.figure(figsize=(12, 6))
    plt.plot(history["best_lengths"], label="Legjobb Tour Hossz", color='green')
    plt.plot(history["avg_lengths"], label="Átlagos Tour Hossz", color='orange', linestyle='--')
    plt.title("Konvergencia Görbe (Generációk vs. Tour Hossz)")
    plt.xlabel("Generáció")
    plt.ylabel("Tour Hossz")
    plt.legend()
    plt.grid(True)
    plt.show()
```
